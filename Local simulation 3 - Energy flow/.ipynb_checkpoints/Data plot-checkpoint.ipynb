```python
import json
import matplotlib.pyplot as plt

# Importing data
filename = '3_spins-15_steps-30_trajectories-dt=0.5.json'
with open(filename, 'r') as file:
    data = json.load(file)

# Collecting interaction edges
spin_pairs = list(range(len(data['Expectation values']['step 0'])))

# plotting energy flow
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(1, 5)

ax1.scatter(spin_pairs, data['Expectation values']['step 0'])
ax2.scatter(spin_pairs, data['Expectation values']['step 1'])
ax3.scatter(spin_pairs, data['Expectation values']['step 2'])
ax4.scatter(spin_pairs, data['Expectation values']['step 3'])
ax5.scatter(spin_pairs, data['Expectation values']['step 4'])

plt.show
```
